## Notes Frequency
Reference: https://pages.mtu.edu/~suits/notefreqs.html

```python
import pandas as pd
import numpy as np
```

```python
notes = pd.read_csv('ScaleFreqs2.csv', sep=',', names=['Note','Hz'], index_col='Note')

notes['Clock Cycles'] =88000/notes.Hz
notes['Clock Cycles Rounded for 88kHz'] = round(notes['Clock Cycles'],0).astype('int64')
notes.loc[notes.Hz <=notes.at['C2','Hz'], 'Clock Cycles Rounded for 88kHz'] = 32*round(notes['Clock Cycles']/32).astype('int64')
notes['Error']= (notes['Clock Cycles']-notes['Clock Cycles Rounded for 88kHz'])*100/notes['Clock Cycles']

# print(max(abs(notes['Error'])))
# print(notes)
```

```python
simple_notes = notes[['Hz', 'Clock Cycles Rounded for 88kHz']].copy()
simple_notes
```

```python
simple_notes.to_csv('Notes_to_Cycles.csv', sep='\t')
```
